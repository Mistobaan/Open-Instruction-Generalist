```python
import simplejson as json
import glob
import os
import tqdm
import gzip
DATASET_PATH = os.path.expanduser('~/workspace/OIG')

files = list(glob.glob(os.path.join(DATASET_PATH, '*p3.jsonl.gz')))

for f in tqdm.tqdm(files):
    print(f)
    with gzip.open(f) as f:
        count = 1
        l = f.readline()
        while l:
            try:
                json.loads(l)
            except json.JSONDecodeError:
                print('error at line %d in file %s' %(count, f))
            l = f.readline()
            count += 1
```
